```python
import cv2
import tensorflow as tf 

CATEGORIES = ["10.000","20.000"]

width, height = 300, 160

def getone(file_path):
    img_array = cv2.imread(os.path.join(file_path), cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (width, height), 1)
    return new_array.reshape(-1, width, height, 1)

def prepare(file_path, img):
    img_array = cv2.imread(os.path.join(file_path, img), cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (width, height), 1)
    return new_array.reshape(-1, width, height, 1)
        

model = tf.keras.models.load_model("MoneyDetect2.0.model")
```

```python
prediction = model.predict([getone("D:\PROGRAMING\DataSets\Money\\10.000\\2a198407-08ef-4b00-a94d-be3bb9d6aefa.jpg")])
print(CATEGORIES[int(prediction[0][0])])
```

```python
prediction = model.predict([getone("D:\PROGRAMING\DataSets\Money\\money.jpg")])
print(CATEGORIES[int(prediction[0][0])])
```

```python
DATADIR = "D:/PROGRAMING/DataSets/Money"

def trainingData():
    for category in CATEGORIES:
        path = os.path.join(DATADIR, category)
        sepulu = 0
        wapulu = 0
        se = 10.000
        wa = 20.000
        for img in os.listdir(path):
            prediction = model.predict([prepare(path, img)])
            hasil = CATEGORIES[int(prediction[0][0])]
            if hasil == "10.000":
                sepulu += 1
                print(sepulu , se)
            elif hasil == "20.000":
                wapulu += 1
                print(wapulu , wa)
        print("-----------------------------------------------")
            
            
            
trainingData()
```
